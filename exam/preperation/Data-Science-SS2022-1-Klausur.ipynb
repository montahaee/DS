(3) Bestimmen Sie mithilfe von K-Means $K=3$ Cluster und visualisieren Sie sie in einem farbkodierten Scatterplot. <span style="color:red;font-weight:bold">\[ /4\] Punkte</span>

- Der Scatterplot zeigt die Datenpunkte farbkodiert nach ihrer Clusterzugehörigkeit, die über K-Means mit $K=3$ ermittelt wurden. (4 Punkte)


```python
kwargs = {'edgecolor': "red",  # for edge color
          'linewidth': 2,  # line width of spot
          }
```

```python
def rand_assignment(data, ks):
    set_size =  data.shape[0]
    return np.random.randint(0, ks, set_size).reshape(set_size, 1)

def get_centroids(data, parts, ks):
    return np.concatenate([np.sum((parts == k) * data, axis=0) /
                         np.sum(k == parts)  for k in range(ks)], axis=0).reshape(ks, data.shape[1])
# return a central radius to make boundaries of random number
def get_boundary(data):
    return int(np.max((np.abs([np.mean(data[:, 1]), np.mean(data[:, 0])]))) + 1)

def navigate_cluster(data, cs, ks):
    # lengths = np.concatenate(np.array([np.sqrt(np.sum((data - cs[k,:]) ** 2, axis=1))
    #                                  for k in range(ks)]) ,axis=1).reshape(data.shape[0], ks)
    # lengths = np.stack(np.array([np.sqrt(np.sum((data - cs[k,:]) ** 2, axis=1))
    lengths = np.stack([np.sqrt(np.sum((data - cs[k,:]) ** 2, axis=1))
                                     for k in range(ks)] ,axis=1)
    next_parts = np.argmin(lengths, axis=1).reshape(data.shape[0],1)
    return lengths, next_parts

# print(get_boundary(X))
# check point of two clusters from a data
def coincided(data, part1, part2):
    return data.shape[0] == np.sum(part1 == part2)


# return randomly initiated centroids
def get_init_centroids(data, ks):
    return np.random.randint(low=-get_boundary(data),
                             high=get_boundary(data), size=(ks, data.shape[1]))


#"naive k-means" as standard algorithm
def naive_k_means(data, ks):
    np.random.seed(4)
    cns = get_init_centroids(data, ks)
    first = rand_assignment(data, ks)
    second = rand_assignment(data, ks)
    sbn.scatterplot(x=cns[:, 0], y=cns[:, 1], c=range(ks),
                    alpha=0.7, s=200, hue=range(ks), palette='pastel', **kwargs)
    while not coincided(data=data, part1=first, part2=second):
        first = second
        cns = get_centroids(data, second, ks)
        lengths, second = navigate_cluster(data, cns, ks)
        sbn.scatterplot(x=cns[:, 0], y=cns[:, 1], c=range(ks),
                        alpha=0.55, s=200, palette='pastel', **kwargs)
    return cns, second
```

(8) Visualisieren Sie den **mittleren Silhouettenindex** (auch *silhouette score* genannt) als Funktion der Clusteranzahl $k\in\{2,\ldots,15 \}$ für das Verfahren aus Teilschritt (5) auf den Daten. <span style="color:red;font-weight:bold">\[ /5\] Punkte</span>

- Ihre Abbildung zeigt den Silhouette Score als Funktion der Clusteranzahl $k\in\{2,\ldots,15 \}$. (4 Punkte)

```python
### FÜHREN SIE DIESE ZELLE AUS BEVOR SIE FORTFAHREN, ABER ÄNDERN SIE HIER KEINEN CODE ###
### EXECUTE THIS CELL BEFORE YOU CONTINUE, BUT DO NOT WRITE ANYTHING INTO IT ###

import numpy as np
from scipy.spatial.distance import cdist
from itertools import combinations

def get_comembership_matrix(X_train, clust_train, X_val, clust_val):
    """
    input:
        X_train, X_val: training and validation set
        clust_train, clust_val: vectors containing indices of cluster memberships in the training and the validation dataset, respectively.
        
    output:
        M: comembership matrix
    """
    # determine centroids
    centroids = np.array([np.mean(X_train[clust_train==k, :], axis=0) for k in np.unique(clust_train)])
    
    # allocate memory to store the co-membership matrix
    M = np.zeros((X_val.shape[0], X_val.shape[0]))
    
    # determine Euclidean distances between test set and centroids from training set
    dist = cdist(X_val, centroids, 'euclidean')
    
    # look up cluster membership by determining minimum distance to centroid
    members = np.argmin(dist, axis=1)

    # populate co-membership matrix
    for i in range(centroids.shape[0]):
        idx = np.argwhere(members == i).squeeze()  # retrieve all members of cluster i
        if idx.size <= 1: continue
        comb = [c for c in combinations(idx, 2)]  # get all pairs of data points belonging to cluster i
        M[tuple(np.transpose(comb))] = 1  # populate upper triangular matrix
  
    # make M symmetric (out of completeness)
    M += M.T  # copy values from upper to lower triangular matrix

    return M


def get_prediction_strength(M, clust_val, k):
    """
    input:
        M: comembership matrix
        clust_val: vector cluster indices of the validation set
        k: number of clusters
        
    output:
        ps: prediction strength    
    """
    
    # allocate memory for prediction strength
    ps = np.zeros(k)
    
    for c in np.unique(clust_val):  # iterate over clusters
        # determine which points belong to cluster c
        idx = np.argwhere(clust_val==c)
        if idx.size <= 1: continue
        for i in idx:
            for j in idx:
                if i == j: continue
                ps[c] += M[i, j]
        ps[c] /= (idx.size*(idx.size-1))
    
    return np.min(ps)
```
