````python
# putting the above in a function to be able to plot it for different ranges
# (with thte old host_subplot style)
def visualize_bavarian_proportion(start,end):
    fig, ax= plt.subplots(figsize=(12,5)) # I need this to controll the figsize (in a oo manner at least...)
    host = host_subplot(111,figure=fig) # see https://stackoverflow.com/questions/33396190/what-are-the-parameters-of-host-subplot-in-matplotlibs-mpl-toolkits-axes-grid1 for magic parameters

    gesamtdeutscher_parasit = host.twinx() # I am kinda sorry for that variable name, but I just had to
    # gesamtdeutscher_parasit.axis["right"].toggle(all=True) # I don't this this does anything...


    host.plot(bavarian_proportion.loc[start:end],label="bayrischer Anteil")
    host.set(
        xlabel="Meldedatum",
        ylabel="prop of bav. cases"
    )

    gesamtdeutscher_parasit.plot(data["Deutschland"].loc[start:end],label="Deutsche Gesamtzahlen")
    gesamtdeutscher_parasit.set(
        ylabel="total number of (new) cases"
    )
    
    

    # hiding the unecessary labels (that should not be there in the first place)

    x_axis = ax.axes.get_xaxis()
    x_axis.set_visible(False)
    y_axis = ax.axes.get_yaxis()
    y_axis.set_visible(False)
    
    # because bavaria has a population of 13 mil, and germany of 83, one should assume that bavaria has 13/83 of the cases
    # thus we draw a line there
    host.axhline(13/83,c="r",label="bav. pop. proport.")
    host.legend()
    
visualize_bavarian_proportion("2020","2020-06")
````

$$
\text{mein Wert}=
\frac{
\frac{c_s}{\bar{c}}-\frac{p_s}{\bar{p}}
}{
\frac{p_s}{\bar{p}}
}
=
\frac{i_s}{\bar{i}}-1
$$
mit $i$ als Inzidienz, $c$ als anzahl der Coronafälle, $p$ als Bevölkerungszahl und $x_s$ jeweils als Wert für den Staat $s$ und $\bar{x}$ als Wert für ganz Deutschland

natürlich gilt:
$$
i_s=\frac{c_s}{p_s}\cdot 10^5
$$

Ok, so we learned about two different clustering methods: K-means vs hierarchical clusteranalysis (HCA)
ersteres arbeitet über Schwerpunkt, zweiteres über dendogramm (baum struktur)
-> bei beiden gibt man aber die Anzahl der Cluster vor

Bei HCA ist die Art der Linkage noch extrem wichtig -> wie wird entschieden welche zwei cluster im nächsten Schritt gemergt werden müssen (genauer: wie wird der Unterschied zwischen zwei Clustern errechnet -> größte Distanz? kleinste Distanz? mittlere Distanz zwischen den Punkten?)

Nicht zu verwechseln mit der dissimilarity measure -> z.B. euklidische Distanz (der Klassiker) aber auch manhatten distanz, **Korrelation** und andere Denkbar.




worauf wir laut Prof beim clustering achten soll
1. Wahl des Unterschiedlichkeitsmaß/dissimilarity measure
2. Vorverarbeitung, z.B. Skalierung
3. Wahl der Klusteranzahl und anderer freier Parameter (z.B. linkage bei HCA)
4. Cluster-Analyse nicht als endpunkt sehen -> bei interpration vorsichtig sein

```
linkage : {'ward', 'complete', 'average', 'single'}, default='ward'
 |      Which linkage criterion to use. The linkage criterion determines which
 |      distance to use between sets of observation. The algorithm will merge
 |      the pairs of cluster that minimize this criterion.
 |  
 |      - 'ward' minimizes the variance of the clusters being merged.
 |      - 'average' uses the average of the distances of each observation of
 |        the two sets.
 |      - 'complete' or 'maximum' linkage uses the maximum distances between
 |        all observations of the two sets.
 |      - 'single' uses the minimum of the distances between all observations
 |        of the two sets.
 |  
```
